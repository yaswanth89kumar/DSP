```python
import pandas as pd
nyc311 = pd.read_csv("data/311_Service_Requests_from_2010_to_Present.csv", low_memory = False)
nyc311_copy = nyc311.copy()
nyc311_copy
```

### Print column names

```python
nyc311_copy.columns
```

### Printing column names which contains null values

```python
null_columns=nyc311_copy.columns[nyc311_copy.isnull().any()]
null_columns
```

### Take a look of null values count for each available column

```python
nyc311_copy[null_columns].isnull().sum()
```

### Droping useless columns 

```python
nyc311_copy.drop(['School or Citywide Complaint', 'Vehicle Type',
       'Taxi Company Borough', 'Taxi Pick Up Location','Garage Lot Name','Ferry Direction','Ferry Terminal Name'], axis = 1, inplace=True)
```

```python
nyc311_copy
```

```python
import numpy as np
```

### Convert Created Date as time stamp

```python
create = pd.to_datetime(nyc311_copy['Created Date']).values
```

```python
create
```

### Convert Closed Date as time stamp

```python
closed = pd.to_datetime(nyc311_copy['Closed Date']).values
```

```python
closed
```

### Calculate Elapsed Time

```python
elapsed = closed - create
elapsed
```

### Converting into seconds

```python
elapsed.astype('timedelta64[s]')
```

### Get the position of Closed Date in the dataframe

```python
closeddate_position = nyc311_copy.columns.get_loc("Closed Date")
closeddate_position
```

### inserting new columns create, close and elapsed in our nyc311_copy dataset

```python
nyc311_copy.insert(closeddate_position + 1, "Create", create)
```

```python
nyc311_copy.insert(closeddate_position + 2, "Close", closed)
```

```python
nyc311_copy.insert(closeddate_position + 3, "Elapsed", elapsed)
```
