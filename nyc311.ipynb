### Convert Created Date as time stamp

```python
create = pd.to_datetime(nyc311_copy['Created Date']).values
```

### Convert Closed Date as time stamp

```python
closed = pd.to_datetime(nyc311_copy['Closed Date']).values
```

```python
closed
```

### Calculate Elapsed Time

```python
elapsed = closed - create
elapsed
```

### Converting into seconds

```python
elapsed.astype('timedelta64[s]')
```

#### Analysis based on hourly bases?
##### How many requests got closed in hourly basis

```python
import matplotlib.pyplot as plt
```

```python
elapsed_time_in_days = elapsed.astype('timedelta64[D]')
```

```python
elapsed_time_in_days = elapsed_time_in_days + 1
```

```python
nyc311_copy['Response_in_days'] = elapsed_time_in_days
```

```python
nyc311_copy
```

```python
Request_elapsed_time = nyc311_copy.Response_in_days.value_counts().sort_index()
Request_elapsed_time
```

##### Requests that has been closed in one week

```python
Request_elapsed_time_head = Request_elapsed_time.head(7)
```

```python
Request_elapsed_time_head.plot.bar(Request_elapsed_time_head.index, Request_elapsed_time_head.values, figsize=(10,6))
plt.xlabel("Elapsed days")
plt.ylabel("Requests closed")
plt.title("Closed Requests Analysys")
for i in range(Request_elapsed_time_head.size):
    plt.annotate(s=Request_elapsed_time_head[i], xy=(i, Request_elapsed_time_head[i]), size = 10, color = "black", weight = "bold", ha="center")    
plt.show()
```

##### Requester closed after one week

```python
Request_elapsed_time_last = Request_elapsed_time.tail(6)
```

```python
Request_elapsed_time_last.plot.bar(Request_elapsed_time_last.index, Request_elapsed_time_last.values, figsize=(10,6))
plt.xlabel("Elapsed days")
plt.ylabel("Requests closed")
plt.title("Closed Requests Analysys")
for i in range(Request_elapsed_time_last.size):
    plt.annotate(s=Request_elapsed_time_last[i], xy=(i, Request_elapsed_time_last[i]), size = 10, color = "black", weight = "bold", ha="center")    
plt.show()
```

##### Analysis Based on Complaint types

```python
Complaint_type_analysis = nyc311_copy["Complaint Type"].value_counts()
Complaint_type_analysis
```

```python
Complaint_type_analysis.plot.bar(Complaint_type_analysis.index, Complaint_type_analysis.values, figsize=(20,6))
plt.xlabel("Complaint Types")
plt.ylabel("Requests closed")
plt.title("Complaint Types Analysys")
for i in range(Complaint_type_analysis.size):
    plt.annotate(s=Complaint_type_analysis[i], xy=(i, Complaint_type_analysis[i]), size = 10, color = "black", weight = "bold", ha="center")    
plt.show()
```

```python
# Data to plot
labels = Complaint_type_analysis.index
sizes = Complaint_type_analysis.values

patches, texts = plt.pie(sizes, shadow=True, startangle=140,wedgeprops   = { 'linewidth' : 1, 
                                'edgecolor' : "black" })
plt.legend(patches, labels, loc="best")

plt.axis('equal')
plt.tight_layout()
plt.show()
```
